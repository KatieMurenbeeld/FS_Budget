```python
budget_data['Average'] = budget_data['2013'] + budget_data['2014'] + budget_data['2015'] + budget_data['2016'] + budget_data['2017'] + budget_data['2018'] / 6
budget_data
```

```python
budget_data_r1 = budget_data[budget_data['Region'] == 1].copy().set_index('Unit Name')
budget_data_r2 = budget_data[budget_data['Region'] == 2].copy().set_index('Unit Name')
budget_data_r3 = budget_data[budget_data['Region'] == 3].copy().set_index('Unit Name')
budget_data_r4 = budget_data[budget_data['Region'] == 4].copy().set_index('Unit Name')
budget_data_r5 = budget_data[budget_data['Region'] == 5].copy().set_index('Unit Name')
budget_data_r6 = budget_data[budget_data['Region'] == 6].copy().set_index('Unit Name')
budget_data_r8 = budget_data[budget_data['Region'] == 8].copy().set_index('Unit Name')
budget_data_r9 = budget_data[budget_data['Region'] == 9].copy().set_index('Unit Name')
budget_data_r10 = budget_data[budget_data['Region'] == 10].copy().set_index('Unit Name')
```
